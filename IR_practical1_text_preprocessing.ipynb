```python
# Import required libraries
import nltk
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
from nltk.stem import PorterStemmer, WordNetLemmatizer
import re

# Download necessary NLTK data (run only once)
nltk.download('punkt')
nltk.download('punkt_tab') 
nltk.download('stopwords')
nltk.download('wordnet')
```

```python
# Sample text document
text = """Text preprocessing is a crucial step in Natural Language Processing (NLP).
It involves cleaning, tokenizing, and preparing the text data for analysis or model training!"""
print("Original Text:\n", text)
```

```python
# Step 1: Lowercasing
text = text.lower()
print("After Lowercasing:\n", text)
```

```python
# Step 2: Remove special characters, numbers, punctuation
text = re.sub(r'[^a-z\s]', '', text)
print("After Removing Special Characters:\n", text)
```

```python
from nltk.tokenize import word_tokenize

tokens = word_tokenize(text)
print("Tokens:\n", tokens)
```

```python
# Step 4: Stop word removal
stop_words = set(stopwords.words('english'))
filtered_tokens = [word for word in tokens if word not in stop_words]
print("After Stop Word Removal:\n", filtered_tokens)
```

```python
# Step 5: Stemming
stemmer = PorterStemmer()
stemmed_tokens = [stemmer.stem(word) for word in filtered_tokens]
print("After Stemming:\n", stemmed_tokens)
```

```python
# Step 6: Lemmatization
lemmatizer = WordNetLemmatizer()
lemmatized_tokens = [lemmatizer.lemmatize(word) for word in stemmed_tokens]
print("After Lemmatization:\n", lemmatized_tokens)
```

```python
# Step 7: Join back into a single string
preprocessed_text = ' '.join(lemmatized_tokens)
print("Final Preprocessed Text:\n", preprocessed_text)
```
